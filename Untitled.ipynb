## Helper functions ##

```python
#Return the note for a pair movie, user
def get_y(movie, person):
    return df[np.logical_and(df.movies == movie, df.people == person)].values[0][0]


#Set prediction as the prediction for the pair movies, people
def set_predictions(movies, people, prediction):
        index = toPredict[np.logical_and(toPredict.movies == movies, toPredict.people == people)].index.values.astype(int)[0]
        predictions.at[index, 'Prediction'] = prediction

def get_rated_movies(person):
    return df[df.people == person].movies.values

def get_movies_to_rate(person):
    return toPredict[toPredict.people == person].movies.values
```

```python
#Get average note for people
def getAverage(people, movies = df) :
    if df.equals(movies):
        return  df.groupby(['people'])['Prediction'].mean()[people]
    else:
        avg = [df[np.logical_and(df.movies == x, df.people == people)].Prediction.values.astype(int)[0] for x in movies]
        if len(avg) == 0 :
            return 0
        else:
            return sum(avg)*1.0/len(avg)
```

```python

def similarity(p1, p2):
    m1 = df[df.people == p1].movies.values
    m2 = df[df.people == p2].movies.values
    common_movies = list(set(m1).intersection(m2))
    p1_avg = getAverage(p1, common_movies)
    p2_avg = getAverage(p2, common_movies)
    p1_note = [df[np.logical_and(df.movies == x, df.people == p1)].Prediction.values.astype(int)[0] for x in common_movies]
    p2_note = [df[np.logical_and(df.movies == x, df.people == p2)].Prediction.values.astype(int)[0] for x in common_movies]
    frac_sup = 0
    frac_inf = 0
    t2 = datetime.datetime.now()
    for i in range(len(p1_note)):
        biais1 = (p1_note[i] - p1_avg)
        biais2 = (p2_note[i] - p2_avg)
        frac_sup += biais1 * biais2
        frac_inf += biais1**2 * biais2**2
    if frac_inf == 0 :
        return 0
    else:
        t3 = datetime.datetime.now()
        return frac_sup/math.sqrt(frac_inf)
```

```python
def prediction(people, movie):
    avg = getAverage(people)
    frac_sup = 0
    frac_inf = 0
    have_rated_movie = df[df.movies == movie].people.values
    for user in set(have_rated_movie):
        frac_sup += get_y(movie, user)*similarity(people, user)
        frac_inf += similarity(people, user)
    return avg + frac_sup/frac_inf
```

```python
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)

from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split

# Load the movielens-100k dataset  UserID::MovieID::Rating::Timestamp
data = Dataset.load_builtin('ml-100k')
trainset, testset = train_test_split(data, test_size=.15)

# Use user_based true/false to switch between user-based or item-based collaborative filtering
algo = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': True})
algo.fit(trainset)

# we can now query for specific predicions
uid = str(196)  # raw user id
iid = str(302)  # raw item id

# get a prediction for specific users and items.
pred = algo.predict(uid, iid, r_ui=4, verbose=True)

# run the trained model against the testset
test_pred = algo.test(testset)

# get RMSE
print("User-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)

# if you wanted to evaluate on the trainset
print("User-based Model : Training Set")
train_pred = algo.test(trainset.build_testset())
accuracy.rmse(train_pred)
```
